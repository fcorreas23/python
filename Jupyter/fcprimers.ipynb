```python
import pandas as pd
```

```python
df1 = pd.read_csv('/home/fcorreas/Desktop/PSEUDOMONAS/Orthogroups.GeneCount.tsv', delimiter="\t")
df2 = pd.read_csv('/home/fcorreas/Desktop/PSEUDOMONAS/filogrupos.txt', delimiter="\t")
orthogroups = df1.set_index("Orthogroup")
phylogroups = df2.groupby("FG")['STRAIN'].apply(list).to_dict()
phylogroups_list = list(phylogroups.keys())
unique = 1 / len(phylogroups)
```

```python
phylogroups_alelos = {}
dataframes = {}
for phylo in phylogroups_list:
    phylogroup = orthogroups.loc[:, phylogroups[phylo]]
    means = phylogroup.loc[:,phylogroups[phylo]].mean(axis=1)
    phylogroup['Means'] = means
    phylogroups_alelos[phylo] = means
    #dataframes[phylo] = phylogroup
    phylogroup.to_csv(f'{phylo}.tsv', sep ='\t' )
```

```python
df_phylogroups = pd.DataFrame(phylogroups_alelos)
df_phylogroups['MEAN'] = df_phylogroups.mean(axis=1)
for x in phylogroups_list:
    #print(len(df_phylogroups[(df_phylogroups[x] == 1) & (df_phylogroups['MEAN'] == unique)].index.tolist()))
    print(df_phylogroups[(df_phylogroups[x] == 1) & (df_phylogroups['MEAN'] == unique)].index.tolist()) 
```
